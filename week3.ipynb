```python
import numpy as np
```

```python
x = np.linspace(-10,10)
y = 1 / (1 + np.exp(-x))
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
plt.style.use('ggplot')
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Logistic graph')
```

```python
tanh = np.tanh(x)
%matplotlib inline
from matplotlib import pyplot as plt
plt.style.use('ggplot')
plt.plot(tanh)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Hyperbolic Tangent graph')
```
